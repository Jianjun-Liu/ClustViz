# Random Forest Regressor

```python
X = df.drop(["value","time"], 1)

y = df.value
```

```python
from sklearn.ensemble import RandomForestRegressor

rnd_clf = RandomForestRegressor(n_estimators=500, random_state=42)

rnd_clf.fit(X, y)

for name, importance in zip(df.columns[1:-1], rnd_clf.feature_importances_):
    print(name, "=", importance)
```

```python
features = df.columns[1:-1]
importances = rnd_clf.feature_importances_
indices = np.argsort(importances)

plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
rnd_clf.score(X, y)
```

# GAM

```python
X.head()
```

```python
from pygam import PoissonGAM, LinearGAM, GAM, s, f, l

#gam = LinearGAM(s(0) + l(1) + s(2) + l(3) + s(4) + s(5))
gam = LinearGAM(s(0) + f(1) + s(2) + f(3) + f(4) + f(5))
gam.gridsearch(X.values, y.values)
```

```python
fig, axs = plt.subplots(1,6, figsize=(16,5));

titles = df.columns[1:-1]
for i, ax in enumerate(axs):
    try:
        XX = gam.generate_X_grid(term=i)
        ax.plot(XX[:, i], gam.partial_dependence(term=i, X=XX))
        ax.plot(XX[:, i], gam.partial_dependence(term=i, X=XX, width=.95)[1], c='r', ls='--')
        ax.set_title(titles[i], fontsize=10);
    except:
        print("error")
        continue
plt.show()
```

```python
gam.summary()
```

```python
diff_pred = abs(gam.predict(X) - y)
```
