```python
varied = make_blobs(n_samples=20,
                             cluster_std=[1.0, 1.0, 1.5],
                             random_state=42)[0]
plt.scatter(varied[:,0], varied[:,1])
#plt.gcf().gca().add_artist(plt.Circle((-5, 0), 5, color="red", fill=False, linewidth=3, alpha=0.7))
plt.show()
```

```python
from pyclustering.cluster.birch import birch, measurement_type
from pyclustering.cluster import cluster_visualizer
from pyclustering.utils import read_sample
from pyclustering.samples.definitions import FAMOUS_SAMPLES
# Sample for cluster analysis (represented by list)
#sample = read_sample(FAMOUS_SAMPLES.SAMPLE_OLD_FAITHFUL)
# Create BIRCH algorithm
birch_instance = birch(varied.tolist(), 3, initial_diameter=3, max_node_entries=4)
# Cluster analysis AVERAGE_INTRA_CLUSTER_DISTANCE = 3

birch_instance.process()
# Obtain results of clustering
clusters = birch_instance.get_clusters()
# Visualize allocated clusters
visualizer = cluster_visualizer()
visualizer.append_clusters(clusters, varied.tolist())
visualizer.show()
```
